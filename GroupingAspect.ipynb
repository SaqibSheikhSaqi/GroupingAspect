```python
from nltk.corpus import wordnet as wn
from nltk.corpus import sentiwordnet
from nltk.tag import pos_tag # doctest: +SKIP
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet # Load the wordnet corpus
import xlsxwriter

#stop_words="this being as we have wher been has had is the . \n on in of from to I we it there for their our and due a this that about through or may be an by etc can also these"
stop_words=["with","any", "am", "are", "would","when","what", "was", "were", "has", "other","this","being", "as", "we", "have", "where", "been", "has", "had", "is","the",".","\n","on","in","of","from","to","I","we","it","there","for","their","our","and","due","a","this","that","about","through","or","may","be","an","by","etc","can","also","these"]
```

```python
def getCatagoryAspect(WORD,r):
    WORD = wn.synsets(WORD)
    df1=""
    Ttag=""
    nam=""
    de=""
    for w in WORD:
        #print w.name()
        #print "Definition:", w.definition()
        nam=nam+w.name() + " "
        de=de+w.definition() + " "
        splitw= w.definition().split()
        for swd in splitw:
            if swd not in stop_words:
                df1=df1+ swd + " "
    
    #print df1
    worksheet2.write(r, 1, nam)
    worksheet2.write(r, 3, de)
    #worksheet.write(r, 3, " ")
    print ("All Tags")
    Ttag=pos_tag(df1.split())
    #print Ttag
    df1=""
    for T in Ttag:
        if T[1]=="JJ" or T[1]=="NN" or T[1]=="JJ":
            df1=df1 + T[0].lower() + " "
    print ("NN and JJ Tags")
    #print pos_tag(df1.split())
    defMatSet.append((set(df1.split())))
    worksheet2.write(r, 4, str(set(df1.split())))
   # worksheet.write(r, 4, " ")
```

```python
workbook = xlsxwriter.Workbook('GroupedAspect.xlsx')
worksheet = workbook.add_worksheet()
worksheet.write(0, 0, "Words")
worksheet.write(0, 1, "Synonyms")
worksheet.write(0, 2, "Path Distance")
worksheet.write(0, 3, "Definitions")
worksheet.write(0, 4, "SDW as JJ and NN")
worksheet.write(0, 5, "Intersection")
worksheet.write(0, 6, "Total Words")
worksheet.write(0, 7, "Decision Through Synonyme")
worksheet.write(0, 8, "Decision Through Path Distance")
worksheet.write(0, 9, "Decision Through Proposed Method")

worksheet2 = workbook.add_worksheet()
worksheet2.write(1, 0, "Words")
worksheet2.write(1, 1, "Synonyms")
worksheet2.write(1, 2, "Path Distance")
worksheet2.write(1, 3, "Definitions")
worksheet2.write(1, 4, "SDW as JJ and NN")
worksheet2.write(1, 5, "Intersection")
worksheet2.write(1, 6, "Total Words")
worksheet2.write(1, 7, "Decision Through Synonyme")
worksheet2.write(1, 8, "Decision Through Path Distance")
worksheet2.write(1, 9, "Decision Through Proposed Method")

words=[]
```
